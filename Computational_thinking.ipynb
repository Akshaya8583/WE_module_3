The code checks whether a given number is a strong number or not.

```python
def get_last_digit(n):
    """Get the last digit of a number."""
    return n % 10

def remove_last_digit(n):
    """Remove the last digit from a number."""
    return n // 10

# Now we can integrate these functions into the previous code.

def calculate_factorial(digit):
    """Calculate the factorial of a digit."""
    if digit == 0:
        return 1
    else:
        result = 1
        for i in range(1, digit + 1):
            result *= i
        return result

def sum_of_factorials_of_digits(num):
    """Calculate the sum of factorials of digits of a number."""
    temp = num
    sum_of_factorials = 0

    while temp > 0:
        digit = get_last_digit(temp)
        sum_of_factorials += calculate_factorial(digit)
        temp = remove_last_digit(temp)

    return sum_of_factorials

# The rest of the code remains the same as before

def is_strong_number(num):
    """Check if a given number is a strong number."""

    if (sum_of_factorials_of_digits(num) == num):
        print(num, "is a strong number.")
    else:
        print(num, "is not a strong number.")

# Call the function
print(is_strong_number(1))
print(is_strong_number(12))
print(is_strong_number(145))
print(is_strong_number(36))
```
